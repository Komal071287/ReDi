## Task 1: Calculating Descriptive Statistics
- Calculate the mean, median, and standard deviation for the 'Age', 'Fare', and 'SibSp' (number of siblings/spouses aboard) columns.
- Calculate the skewness and kurtosis for the 'Fare' column.
- Interpret what these statistics tell you about the distribution of each variable.

### Hint:
Use `.mean()`, `.median()`, `.std()`, and `.skew()` to calculate the required statistics.


```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis

# Load the Titanic dataset into a DataFrame
df = pd.read_csv('D:\Komal\REDI School\Titanic-Dataset.csv')

# Calculate mean, median, and standard deviation
mean_age = df['Age'].mean()
median_age = df['Age'].median()
std_age = df['Age'].std()

mean_fare = df['Fare'].mean()
median_fare = df['Fare'].median()
std_fare = df['Fare'].std()

mean_sibsp = df['SibSp'].mean()
median_sibsp = df['SibSp'].median()
std_sibsp = df['SibSp'].std()

print(f"Age - Mean: {mean_age}, Median: {median_age}, Std: {std_age}")
print(f"Fare - Mean: {mean_fare}, Median: {median_fare}, Std: {std_fare}")
print(f"SibSp - Mean: {mean_sibsp}, Median: {median_sibsp}, Std: {std_sibsp}")

# Calculate skewness and kurtosis for the 'Fare' column
skew_fare = skew(df['Fare'].dropna())
kurtosis_fare = kurtosis(df['Fare'].dropna())

print(f"Fare - Skewness: {skew_fare}, Kurtosis: {kurtosis_fare}")
```

## Task 2: Creating a Correlation Matrix and Heatmap
- Create a correlation matrix for the 'Age', 'Fare', 'SibSp', and 'Parch' (number of parents/children aboard) columns.
- Visualize the correlation matrix using a heatmap with Seaborn.
- Interpret the strongest and weakest correlations from the heatmap.

### Hint:
Use `.corr()` to create the correlation matrix and `sns.heatmap()` to create the heatmap.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Create a correlation matrix for 'Age', 'Fare', 'SibSp', and 'Parch'
corr_matrix = df[['Age', 'Fare', 'SibSp', 'Parch']].corr()
print(corr_matrix)

# Visualize the correlation matrix using a heatmap with Seaborn
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

## Task 3: Creating Advanced Visualizations
- Create a scatter plot to visualize the relationship between 'Fare' and 'Age', with survival status as the hue.
- Create a box plot showing the distribution of 'Fare' for each passenger class ('Pclass').
- Create a pair plot to explore the relationships between 'Age', 'Fare', 'Pclass', and 'Survived'.

### Hint:
Use Seaborn for the visualizations (e.g., `sns.scatterplot()`, `sns.boxplot()`, `sns.pairplot()`).


```python


# Scatter plot to visualize the relationship between 'Fare' and 'Age', with survival status as the hue
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Fare', y='Age', hue='Survived', data=df, palette='coolwarm')
plt.title('Relationship between Fare and Age, with Survival Status as Hue')
plt.xlabel('Fare')
plt.ylabel('Age')
plt.show()

# Box plot showing the distribution of 'Fare' for each passenger class ('Pclass')
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Fare', data=df, palette='coolwarm')
plt.title('Distribution of Fare for Each Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()

# Pair plot to explore the relationships between 'Age', 'Fare', 'Pclass', and 'Survived'
plt.figure(figsize=(12, 10))
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived', palette='coolwarm')
plt.suptitle('Pair Plot of Age, Fare, Pclass, and Survived', y=1.02)
plt.show()
```
